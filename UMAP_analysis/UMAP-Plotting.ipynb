```python
import matplotlib.pyplot as plt
import numpy as np
```

# Baseline

```python
embedding = np.load("D:/TFM/UMAP/Final_results/Baseline/baseline/embedding.npy")
all_labels = np.load("D:/TFM/UMAP/Final_results/Baseline/baseline/all_labels_domain.npy")
all_features = np.load("D:/TFM/UMAP/Final_results/Baseline/baseline/all_features.npy")

# Assuming `umap_embedding` is a numpy array of shape (num_data_points, 2)
# Assuming `labels` is a numpy array of shape (num_data_points,)

plt.figure(figsize=(12, 8))

plt.scatter(embedding[:, 0], embedding[:, 1], c=all_labels, cmap='Accent', s=5)
plt.gca().set_aspect('equal', 'datalim')

unique_labels = np.unique(all_labels)
#plt.colorbar(boundaries=np.arange(np.max(all_labels)+2)-0.5).set_ticks(np.arange(np.max(all_labels)+1))
 
vmin = all_labels.min() - 0.5
vmax = all_labels.max() + 0.5

# Add a color bar with tick marks for each unique label
cbar = plt.colorbar(boundaries=np.arange(np.max(all_labels)+2)-0.5)  # Create a colorbar with custom boundaries
cbar.set_ticks(np.arange(np.max(all_labels)+1))  # Set custom ticks
cbar.set_ticklabels(['A', 'B', 'C', 'D', 'E'])

# Set aspect ratio and title

# Set aspect ratio and title
plt.title('UMAP projection of the data', fontsize=24);
```
